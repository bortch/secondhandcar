### Target - Price

```python
target_var = 'price'
```

```python
beda.get_histplot(df,target_var)
```

À Remarquer: 

* Distribution presque normale
* Longue queue sur la droite représentant un petit nombre de voiture très chère.

Si nécessaire, la feature pourrait être normalisée en utilisant le `log` de la distribution.

```python
beda.get_histplot(df,target_var,log=True)
```

```python
beda.get_ecdf(df,target_var)
```

On peut remarquer que:
* La moyenne est supérieure à la médiane, la distribution est désaxée.
* Moyenne et médiane restent proches relativement à l'étendue des prix. 
* On peut supposer que la moyenne est influencée par des valeurs élevées.

```python
df[target_var].describe()
```

On pourrait investiguer sur ce qui cause le désaxage de la distribution des prix en regardant la dispersion des features relativement au prix.

Plus la dispersion d'une feature sera grande moins elle aura d'influence sur le prix.

En regardant l'évolution des écart-types de chaque feature dans un espace d'investigation, on peut tenter de voir quelles features jouent un rôle dans ce décentrage.

```python
start=df[target_var].median()
stop=df[target_var].max()
df_std_evolution = beda.get_std_evolution(df,target_var, start=start,stop=stop, num=5)
beda.get_features_as_grid_of_lineplot(df_std_evolution)
```

```python
On remarque que les écart-types ont baisser autour des 60000£ et 100000£
```

```python
pivot_price = df[target_var].max()-(df[target_var].max()-60000)/2.
print(f'prix pivot:{pivot_price}')
```

On pourrait investiguer sur les prix au-delà de 102500.0£. 

Pour cela, on va comparer l'écart-type des voitures *chères* et avec l'écart-type des l'ensemble des voitures du dataset.

```python
numerical_feature = beda.get_numerical_columns(df)
max_price = pivot_price
std_high_price = df[df[target_var]>max_price][numerical_feature].std()
std_price = df[numerical_feature].std()
print(std_price.index)
print(f'Standard deviation for price > {max_price}£:\n{std_high_price}\n\nStandard deviation for all price:\n{std_price}')
```

À remarquer:
1. `year`: L'écart-type est proche de 0. Les voitures dont le prix est plus important sont dans les mêmes années.
1. `mileage`: Comparativement à l'écart-type de l'entièreté du dataset, les voitures les plus chères ont des kilomètrages sensiblement moins diverses.
1. L'écart-type du `mpg` a diminué d'un facteur proche de 6.
1. `tax`: l'écart-type est réduit.

```python
df[df['price']>max_price][numerical_feature].describe()
```

```python
df[df['price']<=max_price][numerical_feature].describe()
```

On voit que le prix des voitures augmentes lorsqu'elles:
1. sont plus récentes
1. ont un kilomètrage plus faible
1. ont un `engine_size` relativement élevé
1. ont un `mpg` bas

## Statistiques

### Corrélations

```python
beda.get_pairplot(df, log=['year', 'price'])
```

Les corrélations les plus facilement observables sont:
* `price` et `year`
* `mileage` et `year`
* `mileage` et `price` (sans doute par transitivité)

On peut supposer une légère corrélation entre `tax` et `engine size` ou encore entre `tax` et `mpg`

```python
beda.get_correlation(df, target = 'price')
```

##### Observations des corrélations avec la target `price`

Corrélation positive par ordre décroissant:

1. `price` et `year` sont positivement corrélés.
    Les voiture plus récentes sont plus chères
1. `price` et `engine_size` sont positivement corrélés. Les voitures avec une taille de moteur plus importante sont plus chères.
1. `price` et `tax` plutôt positivement corrélés. C'est peut-être un effet de bord de la corrélation entre `mpg` et `tax`

Corrélation négative par ordre croissant:

1. `price` et `mileage` sont négativement corrélés.
    Plus les distances parcourues par la voiture sont grandes moins chère sera le prix de revente
1. `price` et `mpg` sont négativement corrélés. On pourrait penser que les voitures avec un meilleur rendement se vendraient plus chères, cette corrélation négative pourrait se justifier par le fait que les moteurs plus gros sont plus cher, ont une consommation plus grande et un rendement plus bas.


```python
beda.get_correlation(df, fields = ['year','mileage','engine_size','tax', 'mpg'])
```

##### Observations des corrélations entre features

Les corrélations positives remarquables sont, par ordre décroissant:

1. `mpg` et `mileage` sont positivement corrélés.
1. `tax` et `engine_size` sont positivement corrélés.

Les corrélations négatives remarquables par ordre croissant:

1. `mileage` et `year` sont négativement corrélés. Ce qui fait sens, à moins de n'avoir jamais utilisé sa voiture, plus l'année est ancienne (petite), plus le nombre de miles parcouru est grand.
1. `mpg` et `tax` sont négativement corrélés. La corrélation peut être renforcée par des effets indirects. Si on postule qu'un moteur plus grand consomme plus par miles parcouru et que la taxe augmente avec la taille du moteur, on peut penser que plus le rendement est grand, moins la taxe est élevée. 
1. `mpg` et `engine_size` sont négativement corrélés.
1. `mpg` et `year` sont négativement corrélés.
